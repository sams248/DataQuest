# Line Charts

The dataset contains the monthly unemployment rate as a CSV from January 1948 to August 2016.

```python
import pandas as pd
```

```python
unrate = pd.read_csv("unrate.csv")
```

```python
unrate.head()
```

```python
# convert DATE column to the datetime type
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
```

```python
unrate['DATE'].describe()
```

```python
# the unemployment rate from January 1948 to December 1948
unrate.head(12)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot()
plt.show()
```

```python
%matplotlib inline
```

```python
# Generate a line chart that visualizes the unemployment rates from 1948
plt.plot(unrate['DATE'][:12],unrate['VALUE'][:12])
plt.show()
```

```python
# rotate the x-axis tick labels by 90 degrees
plt.plot(unrate['DATE'][:12],unrate['VALUE'][:12])
plt.xticks(rotation=90)
plt.show()
```

```python
# add axis labels and a title
plt.plot(unrate['DATE'][:12],unrate['VALUE'][:12])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```
