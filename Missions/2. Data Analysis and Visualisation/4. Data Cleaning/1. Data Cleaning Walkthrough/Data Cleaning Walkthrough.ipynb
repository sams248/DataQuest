### Reading in the Data

- `ap_2010.csv` - Data on AP test results

- `class_size.csv` - Data on class size

- `demographics.csv` - Data on demographics

- `graduation.csv` - Data on graduation outcomes

- `hs_directory.csv` - A directory of high schools

- `sat_results.csv` - Data on SAT scores

- `survey_all.txt` - Data on surveys from all schools

- `survey_d75.txt` - Data on surveys from New York City district 75

```python
# reading in the CSV files

import pandas as pd
data_files = ["ap_2010.csv",
"class_size.csv",
"demographics.csv",
"graduation.csv",
"hs_directory.csv",
"sat_results.csv"]

data = {}
for file in data_files:
    data[file.replace(".csv","")] = pd.read_csv(file)
```

### Reading in the Survey Data

```python
# The files are tab delimited and encoded with Windows-1252 encoding
all_survey = pd.read_csv("survey_all.txt",delimiter="\t",encoding="windows-1252")
d75_survey = pd.read_csv("survey_d75.txt",delimiter="\t",encoding="windows-1252")
survey = pd.concat([all_survey,d75_survey],axis=0)
print(survey.head())
```

### Cleaning Up the Surveys

```python
survey["DBN"]=survey["dbn"]
```

```python
survey = survey.loc[:,["DBN","rr_s", "rr_t", "rr_p", "N_s", "N_t", "N_p", "saf_p_11", "com_p_11", "eng_p_11", "aca_p_11", "saf_t_11", "com_t_11", "eng_t_11", "aca_t_11", "saf_s_11", "com_s_11", "eng_s_11", "aca_s_11", "saf_tot_11", "com_tot_11", "eng_tot_11", "aca_tot_11"]]
```

```python
survey.shape
```

```python
data["survey"] = survey
```

### Inserting DBN Fields

```python
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]
```

```python
def pad(n):
    if len(str(n)) > 1:
        return str(n)
    else:
        return str(n).zfill(2)
```

```python
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad)
```

```python
data["class_size"]["padded_csd"].head()
```

```python
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]
```

```python
data["class_size"].head()
```

### Combining the SAT Scores

```python
type(data["sat_results"]["SAT Math Avg. Score"][1])
```

```python
cols = ["SAT Math Avg. Score","SAT Critical Reading Avg. Score","SAT Writing Avg. Score"]
for col in cols:
    data["sat_results"][col] = pd.to_numeric(data["sat_results"][col],errors="coerce")
```

```python
type(data["sat_results"]["SAT Math Avg. Score"][1])
```

```python
data["sat_results"].head()
```

```python
data["sat_results"]["sat_score"] = data["sat_results"][cols[0]] + data["sat_results"][cols[1]] + data["sat_results"][cols[2]]
```

### Parsing Geographic Coordinates for Schools

```python
data["hs_directory"]["Location 1"].head()
```

```python
import re
```

```python
re.findall("\(.+\)","1110 Boston Road\nBronx, NY 10456\n(40.8276026690005, -73.90447525699966)")
```

```python
['(40.8276026690005, -73.90447525699966)'][0].split(",")[0].replace('(',"")
```

```python
def get_lat(location):
    return re.findall("\(.+\)",location)[0].split(",")[0].replace('(',"")
```

```python
data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(get_lat)
data["hs_directory"]["lat"].head()
```

```python
def get_lon(location):
    return re.findall("\(.+\)",location)[0].split(",")[1].replace(')',"")
```

```python
data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(get_lon)
data["hs_directory"]["lon"].head()
```
