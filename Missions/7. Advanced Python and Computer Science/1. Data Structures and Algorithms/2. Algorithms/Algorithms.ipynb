### Finding data in a data set

### Modular code to perform Linear Search

```python
def player_age(name):
    for index, row in nba.iterrows():
        if row[0] == name:
            return row[2]
    return -1
```

```python
allen_age = player_age("Ray Allen")
allen_age
```

```python
durant_age = player_age("Kevin Durant")
durant_age
```

```python
shaq_age = player_age("Shaquille O'Neal")
shaq_age
```

### Constant time algorithms

```python
# Converting degrees Celcius to degrees Fahrenheit
 
def celcius_to_fahrenheit(degrees):
    step_1 = degrees * 1.8
    step_2 = step_1 + 32
    return step_2
```

```python
# Reversing a list

def reverse(ls):
    length = len(ls)
    new_list = []
    for i in range(length):
        new_list[i] = ls[length - i]
    return new_list

reversing_a_list_complexity = "Not constant!"
```

```python
# Printing a blastoff message after a countdown

def blastoff(message):
    count = 10
    for i in range(count):
        print(count - i)
    print(message)
```

### Some other algorithms

```python
def length(ls):
    count = 0
    for elem in ls:
        count = count + 1
        
length_time_complexity = "linear"
```

```python
def is_empty_1(ls):
    if length(ls) == 0:
        return True
    else:
        return False
    
is_empty_1_complexity = "linear"
```

```python
def is_empty_2(ls):
    for element in ls:
        return False
    return True

is_empty_2_complexity = "constant"
```
